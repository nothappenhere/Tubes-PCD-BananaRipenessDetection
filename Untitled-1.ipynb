```python
import cv2
import matplotlib.pyplot as plt

ref_img = cv2.imread("img/cavendish/Image_12.jpg")
ref_img = cv2.cvtColor(ref_img, cv2.COLOR_BGR2RGB)

many_fruit_img = cv2.imread("img/cavendish/shutterstock_533882299.jpg")
many_fruit_img = cv2.cvtColor(many_fruit_img, cv2.COLOR_BGR2RGB)

fig =  plt.figure(figsize=(16, 16))
plt.subplot(1, 2, 1)
plt.imshow(ref_img)
plt.title("Reference Image")
plt.xticks([])
plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(many_fruit_img)
plt.title("Many Reference Image")
plt.xticks([])
plt.yticks([])
```

```python
gray_ref_img = cv2.cvtColor(ref_img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_ref_img, cmap='gray')
plt.title("GRAYSCALE Reference Image")
plt.xticks([])
plt.yticks([])
```

```python
bin_gray_ref_img = cv2.threshold(gray_ref_img, 245, 255, cv2.THRESH_BINARY)

plt.imshow(bin_gray_ref_img[1], cmap='gray')
plt.title("BINARY Reference Image")
plt.xticks([])
plt.yticks([])
```

```python
inv_bin_gray_ref_img = ~bin_gray_ref_img[1]

plt.imshow(inv_bin_gray_ref_img, cmap='gray')
plt.title("INVERTED BINARY Reference Image")
plt.xticks([])
plt.yticks([])
```

```python
ref_contour_list, hierarchy = cv2.findContours(inv_bin_gray_ref_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

with_contours = cv2.drawContours(ref_img, ref_contour_list, -1, (255, 0, 0), 10)

plt.imshow(ref_img)
plt.title("Reference Image Red Line")
plt.xticks([])
plt.yticks([])
```
